## Step 13: Identify Numerical and Categorical Features

Categorizing features into numerical and categorical groups helps determine the preprocessing techniques required during the next phase, such as encoding categorical variables and scaling numerical features.

```python
# Crop Dataset

crop_numerical = crop_df.select_dtypes(include=["int64", "float64"]).columns.tolist()
crop_categorical = crop_df.select_dtypes(include=["object"]).columns.tolist()

print("Crop Dataset")
print("Numerical Features:", crop_numerical)
print("Categorical Features:", crop_categorical)
```

```python
# Fertilizer Dataset

fert_numerical = fertilizer_df.select_dtypes(include=["int64", "float64"]).columns.tolist()
fert_categorical = fertilizer_df.select_dtypes(include=["object"]).columns.tolist()

print("Fertilizer Dataset")
print("Numerical Features:", fert_numerical)
print("Categorical Features:", fert_categorical)
```
