# Precision Farming Recommendation Engine

## Notebook 02: Data Cleaning & Preprocessing

This notebook prepares the Crop Recommendation and Fertilizer Recommendation datasets for machine learning by validating data quality, transforming categorical variables into numerical representations, scaling numerical features where required, and generating processed datasets for model development.

Since both datasets were found to be complete during the data collection phase, the primary focus of this notebook is preprocessing rather than extensive data cleaning.

## Objective

The objectives of this notebook are to:

- Load the collected datasets.
- Verify data quality.
- Validate numerical features.
- Perform feature selection.
- Encode categorical variables.
- Encode target variables.
- Split the datasets into training and testing sets.
- Scale numerical features.
- Save the processed datasets and preprocessing objects for future use.

## Step 1: Import Required Libraries

Import all libraries required for preprocessing and preparing the datasets for machine learning.

```python
# Import required libraries

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import joblib

# Display all columns
pd.set_option("display.max_columns", None)

print("Libraries imported successfully.")
```

## Step 2: Load the Datasets

Load the Crop Recommendation and Fertilizer Recommendation datasets from the project's raw data directory into Pandas DataFrames. These datasets will be used throughout the preprocessing pipeline.

```python
# Load datasets

crop_df = pd.read_csv("../data/raw/Crop_recommendation.csv")
fertilizer_df = pd.read_csv("../data/raw/fertilizer_recommendation.csv")

print("Crop Dataset Shape       :", crop_df.shape)
print("Fertilizer Dataset Shape :", fertilizer_df.shape)

print("\nDatasets loaded successfully.")
```

## Step 3: Create Working Copies of the Datasets

To preserve the original datasets, create independent working copies that will be used for all preprocessing operations performed in this notebook.

```python
# Create working copies

crop_data = crop_df.copy()
fertilizer_data = fertilizer_df.copy()

print("Working copies created successfully.")
```

## Step 4: Crop Dataset Information

Examine the structure of the Crop Recommendation dataset to verify the number of records, feature names, data types, and non-null values. This helps confirm the dataset is complete and suitable for preprocessing.

```python
# Display dataset information

print("Crop Dataset Information\n")

crop_data.info()
```

## Step 5: Missing Value Analysis

Identify missing values in the Crop Recommendation dataset. Missing values can negatively affect machine learning models and must be handled appropriately before training.

```python
# Check missing values

missing_values = crop_data.isnull().sum()

print("Missing Values in Crop Dataset\n")
display(missing_values)

print("\nTotal Missing Values :", missing_values.sum())
```

## Step 6: Duplicate Record Analysis

Check for duplicate records in the Crop Recommendation dataset. Duplicate observations can introduce bias during model training and should be identified before preprocessing.

```python
# Check duplicate records

duplicate_count = crop_data.duplicated().sum()

print("Duplicate Records :", duplicate_count)
```

## Step 7: Validate Numerical Features

Inspect the numerical features of the Crop Recommendation dataset to verify that the recorded values fall within expected ranges and to identify any obvious inconsistencies or invalid values.

```python
# Summary statistics for numerical features

display(crop_data.describe())
```

## Step 8: Identify Feature Data Types

Categorize the dataset features into numerical and categorical variables. This helps determine the appropriate preprocessing techniques, such as encoding categorical variables and scaling numerical features.

```python
# Identify numerical and categorical features

numerical_features = crop_data.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_features = crop_data.select_dtypes(include=["object"]).columns.tolist()

print("Numerical Features:")
print(numerical_features)

print("\nCategorical Features:")
print(categorical_features)
```

## Step 9: Feature Selection

Separate the independent features (X) from the target variable (y). The input features will be used for model training, while the target variable represents the crop to be predicted.

```python
# Separate features and target

X_crop = crop_data.drop(columns=["label"])
y_crop = crop_data["label"]

print("Feature Matrix Shape :", X_crop.shape)
print("Target Vector Shape  :", y_crop.shape)
```

## Step 10: Encode the Target Variable

Encode the target variable (`label`) into numerical values using `LabelEncoder`. This transformation enables machine learning algorithms to work with the categorical crop labels.

```python
# Encode target variable

crop_label_encoder = LabelEncoder()

y_crop_encoded = crop_label_encoder.fit_transform(y_crop)

print("Target variable encoded successfully.")
print("\nNumber of Classes:", len(crop_label_encoder.classes_))
```

## Step 11: Display Label Encoding Mapping

Display the mapping between the original crop names and their corresponding encoded values. This mapping will be useful during prediction and model deployment.

```python
# Create label mapping

crop_label_mapping = pd.DataFrame({
    "Encoded Label": range(len(crop_label_encoder.classes_)),
    "Crop": crop_label_encoder.classes_
})

display(crop_label_mapping)
```

## Step 12: Split the Dataset into Training and Testing Sets

Split the Crop Recommendation dataset into training and testing sets. The training set will be used to train the machine learning models, while the testing set will be reserved for evaluating their performance on unseen data.

An 80:20 split is used, and stratification is applied to preserve the distribution of crop classes in both subsets.

```python
# Split dataset

X_crop_train, X_crop_test, y_crop_train, y_crop_test = train_test_split(
    X_crop,
    y_crop_encoded,
    test_size=0.20,
    random_state=42,
    stratify=y_crop_encoded
)

print("Training Features :", X_crop_train.shape)
print("Testing Features  :", X_crop_test.shape)

print()

print("Training Labels   :", y_crop_train.shape)
print("Testing Labels    :", y_crop_test.shape)
```

## Step 13: Feature Scaling

Standardize the numerical input features using `StandardScaler`. Scaling ensures that all features have a similar range, improving the performance of distance-based and gradient-based machine learning algorithms.

```python
# Initialize scaler

crop_scaler = StandardScaler()

# Fit on training data and transform

X_crop_train_scaled = crop_scaler.fit_transform(X_crop_train)

# Transform testing data

X_crop_test_scaled = crop_scaler.transform(X_crop_test)

print("Feature scaling completed successfully.")
```

## Step 14: Convert Scaled Data Back to DataFrames

Convert the scaled NumPy arrays back into Pandas DataFrames while preserving the original feature names.

```python
# Convert scaled arrays back to DataFrames

X_crop_train_scaled = pd.DataFrame(
    X_crop_train_scaled,
    columns=X_crop.columns,
    index=X_crop_train.index
)

X_crop_test_scaled = pd.DataFrame(
    X_crop_test_scaled,
    columns=X_crop.columns,
    index=X_crop_test.index
)

print("Scaled datasets converted to DataFrames.")
```

## Step 15: Save Preprocessing Objects

Save the fitted preprocessing objects for future use during model training and deployment. These objects ensure that new input data is processed in the same way as the training data.

The following objects will be saved:

- Crop Label Encoder
- Crop Standard Scaler

```python
# Save preprocessing objects

joblib.dump(crop_label_encoder, "../models/crop_label_encoder.pkl")
joblib.dump(crop_scaler, "../models/crop_scaler.pkl")

print("Crop preprocessing objects saved successfully.")
```

## Step 16: Save Processed Datasets

Save the processed training and testing datasets for future model development. Storing these datasets avoids repeating preprocessing steps and ensures consistency throughout the project.

```python
# Save processed datasets

X_crop_train_scaled.to_csv("../data/processed/X_crop_train.csv", index=False)
X_crop_test_scaled.to_csv("../data/processed/X_crop_test.csv", index=False)

pd.Series(y_crop_train).to_csv("../data/processed/y_crop_train.csv", index=False)
pd.Series(y_crop_test).to_csv("../data/processed/y_crop_test.csv", index=False)

print("Processed crop datasets saved successfully.")
```

## Step 17: Verify Saved Files

Verify that the preprocessing objects and processed datasets have been successfully saved to their respective project directories.

```python
import os

print("Models Directory")
print("-" * 40)

print(os.listdir("../models"))

print("\nProcessed Data Directory")
print("-" * 40)

print(os.listdir("../data/processed"))
```

# Fertilizer Dataset Preprocessing

## Step 18: Dataset Information

Inspect the structure of the Fertilizer Recommendation dataset to verify the number of records, feature names, data types, and non-null values. This confirms that the dataset is complete and ready for preprocessing.

```python
print("Fertilizer Dataset Information\n")

fertilizer_data.info()
```

## Step 19: Missing Value Analysis

Check the Fertilizer Recommendation dataset for missing values. Identifying missing values is an essential preprocessing step to ensure data quality before model training.

```python
missing_values = fertilizer_data.isnull().sum()

print("Missing Values in Fertilizer Dataset\n")

display(missing_values)

print("\nTotal Missing Values :", missing_values.sum())
```

## Step 20: Duplicate Record Analysis

Identify duplicate records in the Fertilizer Recommendation dataset. Duplicate observations can introduce bias into machine learning models and should be detected before further preprocessing.

```python
duplicate_count = fertilizer_data.duplicated().sum()

print("Duplicate Records :", duplicate_count)
```

## Step 21: Validate Numerical Features

Review the statistical summary of the numerical features to ensure that recorded values fall within expected ranges and to identify any potential inconsistencies.

```python
display(fertilizer_data.describe())
```

## Step 22: Identify Numerical and Categorical Features

Separate the fertilizer dataset features into numerical and categorical groups. This helps determine which preprocessing techniques should be applied to each feature type.

```python
# Identify numerical and categorical features

fert_numerical_features = fertilizer_data.select_dtypes(
    include=["int64", "float64"]
).columns.tolist()

fert_categorical_features = fertilizer_data.select_dtypes(
    include=["object"]
).columns.tolist()

print("Numerical Features:")
print(fert_numerical_features)

print()

print("Categorical Features:")
print(fert_categorical_features)
```

## Step 23: Feature Selection

Separate the independent variables (features) from the target variable. The input features will be used to train the fertilizer recommendation model, while the target variable represents the recommended fertilizer.

```python
# Separate features and target

X_fertilizer = fertilizer_data.drop(columns=["Recommended_Fertilizer"])

y_fertilizer = fertilizer_data["Recommended_Fertilizer"]

print("Feature Matrix Shape :", X_fertilizer.shape)
print("Target Vector Shape  :", y_fertilizer.shape)
```

## Step 24: Identify Input Feature Types

Classify the independent variables into numerical and categorical features. Numerical features will be standardized, while categorical features will be encoded into numerical values.

```python
# Input feature types

categorical_columns = X_fertilizer.select_dtypes(include=["object"]).columns.tolist()

numerical_columns = X_fertilizer.select_dtypes(
    include=["int64", "float64"]
).columns.tolist()

print("Categorical Input Features:")
print(categorical_columns)

print()

print("Numerical Input Features:")
print(numerical_columns)
```

## Step 25: Encode Categorical Input Features

Convert all categorical input features into numerical values using separate `LabelEncoder` objects. Each categorical feature is encoded independently to preserve its unique mapping, and the fitted encoders are stored for future inference and deployment.

```python
# Encode categorical input features

feature_encoders = {}

for column in categorical_columns:
    encoder = LabelEncoder()

    X_fertilizer[column] = encoder.fit_transform(X_fertilizer[column])

    feature_encoders[column] = encoder

print("Categorical input features encoded successfully.")

print("\nEncoded Columns:")

for column in feature_encoders.keys():
    print("-", column)
```

## Step 26: Encode the Target Variable

Encode the target variable (`Recommended_Fertilizer`) into numerical labels using `LabelEncoder`. This allows machine learning algorithms to train on categorical fertilizer classes while preserving the mapping for future predictions.

```python
# Encode target variable

fertilizer_label_encoder = LabelEncoder()

y_fertilizer_encoded = fertilizer_label_encoder.fit_transform(y_fertilizer)

print("Target variable encoded successfully.")

print("\nNumber of Fertilizer Classes:",
      len(fertilizer_label_encoder.classes_))
```

## Step 27: Display Target Label Mapping

Display the mapping between fertilizer names and their corresponding encoded values. This mapping will be required when converting model predictions back into fertilizer names during deployment.

```python
# Display fertilizer label mapping

fertilizer_label_mapping = pd.DataFrame({
    "Encoded Label": range(len(fertilizer_label_encoder.classes_)),
    "Recommended Fertilizer": fertilizer_label_encoder.classes_
})

display(fertilizer_label_mapping)
```

## Step 28: Split the Dataset into Training and Testing Sets

Split the fertilizer dataset into training and testing sets using an 80:20 ratio. Stratification is applied to preserve the class distribution of the target variable in both subsets.

```python
# Train-test split

X_fert_train, X_fert_test, y_fert_train, y_fert_test = train_test_split(
    X_fertilizer,
    y_fertilizer_encoded,
    test_size=0.20,
    random_state=42,
    stratify=y_fertilizer_encoded
)

print("Training Features :", X_fert_train.shape)
print("Testing Features  :", X_fert_test.shape)

print()

print("Training Labels   :", y_fert_train.shape)
print("Testing Labels    :", y_fert_test.shape)
```

## Step 29: Scale Numerical Features

Standardize only the numerical input features using `StandardScaler`. Categorical features remain unchanged because they represent encoded categories rather than continuous measurements.

```python
# Initialize scaler

fertilizer_scaler = StandardScaler()

# Scale numerical features

X_fert_train[numerical_columns] = fertilizer_scaler.fit_transform(
    X_fert_train[numerical_columns]
)

X_fert_test[numerical_columns] = fertilizer_scaler.transform(
    X_fert_test[numerical_columns]
)

print("Numerical features scaled successfully.")
```

## Step 30: Save Preprocessing Objects

Save all preprocessing objects including feature encoders, target encoder, and scaler. These objects ensure that future input data is transformed exactly as it was during model training.

```python
import joblib
import os

# Create models directory if it doesn't exist
os.makedirs("../models", exist_ok=True)

# Save feature encoders
joblib.dump(feature_encoders,
            "../models/fertilizer_feature_encoders.pkl")

# Save target encoder
joblib.dump(fertilizer_label_encoder,
            "../models/fertilizer_label_encoder.pkl")

# Save scaler
joblib.dump(fertilizer_scaler,
            "../models/fertilizer_scaler.pkl")

print("Fertilizer preprocessing objects saved successfully.")
```

## Step 31: Save Processed Datasets

Save the processed training and testing datasets. These datasets will be used directly during the machine learning model development phase.

```python
# Create processed data directory
os.makedirs("../data/processed", exist_ok=True)

# Save processed datasets

X_fert_train.to_csv("../data/processed/X_fert_train.csv", index=False)
X_fert_test.to_csv("../data/processed/X_fert_test.csv", index=False)

pd.DataFrame(y_fert_train,
             columns=["Recommended_Fertilizer"]).to_csv(
                 "../data/processed/y_fert_train.csv",
                 index=False
             )

pd.DataFrame(y_fert_test,
             columns=["Recommended_Fertilizer"]).to_csv(
                 "../data/processed/y_fert_test.csv",
                 index=False
             )

print("Processed fertilizer datasets saved successfully.")
```

## Step 32: Verify Saved Files

Verify that all preprocessing objects and processed datasets have been successfully stored in their respective directories.

```python
print("Models Directory")
print("-" * 40)

print(sorted(os.listdir("../models")))

print()

print("Processed Data Directory")
print("-" * 40)

print(sorted(os.listdir("../data/processed")))
```
