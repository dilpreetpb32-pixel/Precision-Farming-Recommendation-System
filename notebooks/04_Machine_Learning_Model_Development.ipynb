# 📂 Phase 4 – Machine Learning Model Development

## 🎯 Objective

Develop and train multiple machine learning models for both the Crop Recommendation and Fertilizer Recommendation systems using the preprocessed datasets. The objective is to learn patterns from soil, environmental, and agricultural features, generate accurate recommendations, and prepare trained models for performance evaluation, comparison, and deployment in the subsequent phases.

# Step 1 – Import Required Libraries

Import all the necessary libraries required for training, managing, and saving machine learning models. These libraries provide classification algorithms, performance evaluation utilities, and model serialization capabilities that will be used throughout the model development phase.

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

# Machine Learning Models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# Save Models
import joblib

print("Libraries imported successfully.")
```

# Step 2 – Load Processed Datasets

Load the preprocessed training and testing datasets created during the preprocessing phase. These datasets have already been cleaned, encoded, scaled (where required), and split into training and testing sets, making them ready for machine learning model development.

```python
# ==========================
# Crop Recommendation Dataset
# ==========================

X_crop_train = pd.read_csv("../data/processed/X_crop_train.csv")
X_crop_test = pd.read_csv("../data/processed/X_crop_test.csv")

y_crop_train = pd.read_csv("../data/processed/y_crop_train.csv").squeeze()
y_crop_test = pd.read_csv("../data/processed/y_crop_test.csv").squeeze()


# ==========================
# Fertilizer Recommendation Dataset
# ==========================

X_fert_train = pd.read_csv("../data/processed/X_fert_train.csv")
X_fert_test = pd.read_csv("../data/processed/X_fert_test.csv")

y_fert_train = pd.read_csv("../data/processed/y_fert_train.csv").squeeze()
y_fert_test = pd.read_csv("../data/processed/y_fert_test.csv").squeeze()


print("Processed datasets loaded successfully.\n")

print("Crop Training Features :", X_crop_train.shape)
print("Crop Testing Features  :", X_crop_test.shape)
print("Crop Training Labels   :", y_crop_train.shape)
print("Crop Testing Labels    :", y_crop_test.shape)

print("\n--------------------------------------\n")

print("Fertilizer Training Features :", X_fert_train.shape)
print("Fertilizer Testing Features  :", X_fert_test.shape)
print("Fertilizer Training Labels   :", y_fert_train.shape)
print("Fertilizer Testing Labels    :", y_fert_test.shape)
```

# Step 3.1 – Crop Recommendation using Logistic Regression

## 🎯 Objective

Train a Logistic Regression classifier using the preprocessed crop recommendation dataset. Logistic Regression serves as a strong baseline model for multi-class classification and helps evaluate how well a simple linear model can distinguish between different crop categories based on soil nutrients and environmental conditions.

```python
# Create the model
crop_logistic_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

# Train the model
crop_logistic_model.fit(X_crop_train, y_crop_train)

print("Crop Logistic Regression model trained successfully.")
```

# Step 3.2 – Crop Recommendation using Decision Tree

## 🎯 Objective

Train a Decision Tree classifier using the preprocessed crop recommendation dataset. Decision Trees learn hierarchical decision rules from the input features, making them effective for capturing non-linear relationships between soil properties, environmental conditions, and crop recommendations.

```python
# Create the model
crop_decision_tree_model = DecisionTreeClassifier(
    random_state=42
)

# Train the model
crop_decision_tree_model.fit(X_crop_train, y_crop_train)

print("Crop Decision Tree model trained successfully.")
```

# Step 3.3 – Crop Recommendation using Random Forest

## 🎯 Objective

Train a Random Forest classifier using the preprocessed crop recommendation dataset. Random Forest combines multiple decision trees to improve prediction accuracy, reduce overfitting, and provide a robust model for crop recommendation.

```python
# Create the model
crop_random_forest_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Train the model
crop_random_forest_model.fit(X_crop_train, y_crop_train)

print("Crop Random Forest model trained successfully.")
```

# Step 3.4 – Crop Recommendation using K-Nearest Neighbors

## 🎯 Objective

Train a K-Nearest Neighbors (KNN) classifier using the preprocessed crop recommendation dataset. KNN classifies crops by identifying the most similar training samples based on feature similarity, making it a useful distance-based classification algorithm.

```python
# Create the model
crop_knn_model = KNeighborsClassifier(
    n_neighbors=5
)

# Train the model
crop_knn_model.fit(X_crop_train, y_crop_train)

print("Crop K-Nearest Neighbors model trained successfully.")
```

# Step 3.5 – Crop Recommendation using Support Vector Machine

## 🎯 Objective

Train a Support Vector Machine (SVM) classifier using the preprocessed crop recommendation dataset. SVM identifies optimal decision boundaries between crop classes, making it effective for handling complex multi-class classification problems.

```python
# Create the model
crop_svm_model = SVC(
    kernel="rbf",
    random_state=42
)

# Train the model
crop_svm_model.fit(X_crop_train, y_crop_train)

print("Crop Support Vector Machine model trained successfully.")
```
